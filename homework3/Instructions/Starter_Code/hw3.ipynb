```python
#rolling window stats
port_rolling_std = port_combined.rolling(window=21).std()
```

```python
#plot rolling std
port_rolling_std.plot(figsize=(14,10), title='21 Day Rolling STD')
```

```python
#calculate correlation
correlation_df = port_combined.corr()
```

```python
#plot correlation
correlation_df
```

```python
#calculate covariance
covariance = port_combined['Algo 1'].rolling(window=21).cov(port_combined['Close'])
```

```python
#calculate variance
variance =  port_combined['Algo 1'].rolling(window=21).var()
```

```python
#calculate Beta
beta = covariance / variance
beta
```

```python
#Plot Beta over time
beta.plot(figsize=(14, 10), title='21 Day Rolling Beta of Algo 1 Fund')
```

```python
#Calculate Exponentially Weighted Average
port_combined.ewm(halflife=21).mean()
```

```python
#Plot of Rollinf EWM
port_combined.ewm(halflife=21).mean().plot(figsize=(14, 10), title='EWM Plot of Combined Portfolio')
```
